```python
import pandas as pd

train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
```

1. 성별, 클라스- 생존율
2. 가족(SibSp,Parch), 성별 생존율

```python
features = ['Pclass', 'Sex', 'Cabin', 'Family']
train_input = pd.get_dummies(train_data[features]) # get_dummies 글자로 된 데이터 숫자로 표현
test_input = pd.get_dummies(test_data[features])
train_input.head()
```

```python
cabin_null = train_data[['Cabin']].isnull()
train_data['Cabin'] = cabin_null
cabin_null_t = test_data[['Cabin']].isnull()
test_data['Cabin'] = cabin_null_t
```

```python
train_data['Family'] = train_data['SibSp'] + train_data['Parch']
test_data['Family'] = test_data['SibSp'] + test_data['Parch']
train_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
train_target = train_data[['Survived']]
```

```python
rfc.fit(train_input, train_target)
```

```python
rfc.feature_importances_
```

```python
from sklearn.model_selection import cross_validate
```

```python
scores = cross_validate(rfc, train_input, train_target, n_jobs=-1)
```

```python
scores
```

```python
predictions = rfc.predict(test_input)
```

```python
result = {
    'PassengerId': test_data['PassengerId'],
    'Survived': predictions,
}
output = pd.DataFrame(result)
```

```python
output.to_csv('my_submission.csv', index=False)
```
