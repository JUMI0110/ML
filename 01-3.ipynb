```python
import pandas as pd
```

```python
df = pd.read_csv('data/Fish.csv')
df.head()
df.shape
```

```python
df['Species'].unique()
```

```python
bream_df = df[df['Species'] == 'Bream']
```

```python
# bream_df
bream_length = bream_df['Length2'].tolist() # 리스트로 형태 변환
bream_weight = bream_df['Weight'].tolist()

# bream_length
```

```python
import matplotlib.pyplot as plt

plt.scatter(bream_length, bream_weight) # 산점도
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
smelt_df = df[df['Species'] == 'Smelt']
# smelt_df.head()
smelt_length = smelt_df['Length2'].tolist() 
smelt_weight = smelt_df['Weight'].tolist()
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]
# fish_data
```

```python
target_data = [1] * len(bream_length) + [0] * len(smelt_length)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier() #최근접이웃 알고리즘 근접하는 이웃의 개수에 따라 데이터 예측 (파라미터 개수 다수결)
```

```python
# 지도학습 (학습할데이터, 정답데이터)를 주고 학습
kn.fit(fish_data, target_data)
```

```python
kn.score(fish_data, target_data) # 1.0 -> 100점
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(25, 400, c='#ff0000', s=100, marker='*')  # scatter()옵션 : c='color' s='size' marker='marker'
plt.scatter(16, 30)
plt.scatter(25, 150, marker='D')
plt.scatter(smelt_length, smelt_weight)
plt.show()
```

```python
kn.predict([[25, 400]])  # 예측명령 
```

```python
kn.predict([[16, 30]])
```

```python
kn.predict([[25, 150]])
```

```python
kn49 = KNeighborsClassifier(49) 
```

```python
kn49.fit(fish_data, target_data)
```

```python
kn49.score(fish_data, target_data)
```

```python
#하이퍼파라미터 파라미터 값에 따라 달라지는 점수 
kn = KNeighborsClassifier()
kn.fit(fish_data, target_data)

for n in range(5, 50):
    kn.n_neighbors = n

    score = kn.score(fish_data, target_data)

    print(n, score)
    if score < 1:
        break
```

## iris 데이터 연습

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

 1. 데이터 가져오기
 2. 어떤 특성을 사용할 지
 3. kn학습
 4. matplotlib 시각화
 5. 예측

```python
df.shape
df['species'].unique()
```

```python
setosa_df = df[df['species'] == 'setosa']
# setosa_df
```

```python
versicolor_df = df[df['species']== 'versicolor']
# versicolor_df
```

```python
virginica_df = df[df['species'] == 'virginica']
# virginica_df
```

```python
se_length = setosa_df['petal_length'].tolist()
se_width = setosa_df['petal_width'].tolist()

ver_length = versicolor_df['petal_length'].tolist()
ver_width = versicolor_df['petal_width'].tolist()

vir_length = virginica_df['petal_length'].tolist()
vir_width = virginica_df['petal_width'].tolist()
```

```python
plt.scatter(se_length, se_width)
plt.scatter(vir_length, vir_width)
# plt.scatter(ver_length, ver_width)
plt.scatter(1, 1)
plt.scatter(4, 1.2)
plt.show()
```

```python
length = se_length + vir_length
width = se_width + vir_width

iris_data = [[l, w] for l, w in zip(length, width)]
target_data = [1] * len(se_length) + [0] * len(vir_length)
```

```python
kn = KNeighborsClassifier()
kn.fit(iris_data, target_data)
```

```python
kn.score(iris_data, target_data)
```

```python
kn.predict([[1, 1]])
```

```python
kn.predict([[4, 1.2]])
```
